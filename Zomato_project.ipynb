# Zomato Data Analysis Project

# Step 1 - Importing Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

# Step 2 - Create the Data Frame

```python
dataframe = pd.read_csv("Zomato data .csv")
```

```python
print(dataframe)
```

```python
dataframe
```

# Convert the Data type of column - rate

```python
def handleRate(value):
    value = str(value).split('/')
    value=value[0];
    return float(value)

dataframe['rate']=dataframe['rate'].apply(handleRate)
print(dataframe.head())
```

```python
dataframe.info()
```

# Qus 1 - What type of restaurant do the majority of customers order from ?

```python
dataframe.head()
```

```python
sns.countplot(x=dataframe['listed_in(type)'],
hue=dataframe['listed_in(type)'],
palette='viridis', legend=False)
plt.xlabel("type of restaurant")
plt.show()
```

 # Conclusion - Majority of the restaurant falls in dining category

# Qus 2 - How many votes has each type of restaurant received from customers?

```python
dataframe.head()
```

```python
grouped_data = dataframe.groupby('listed_in(type)')['votes'].sum()
result = pd.DataFrame({'votes':grouped_data})
plt.plot(result, c="green", marker="o")
plt.xlabel("Type of restaurant", c="black", size=30)
plt.ylabel("Votes", c="black", size=30)
```

# Conclusion - Dining Restaurant has received maximum number of votes

# Qus 3 - What are the ratings that the majority of restaurants have received?

```python
dataframe.head()
```

```python
plt.hist(dataframe['rate'],bins = 10)
plt.title("ratings distribution")
plt.show()
```

# Conclusion - The majority of restaurants received ratings from 3.5 to 4

# Qus 4 - Zomato has observed thst most couples order most of their food online. what is their average spending on each other?

```python
dataframe.head()
```

```python
couple_data=dataframe['approx_cost(for two people)']
sns.countplot(x=couple_data)
```

# Conclusion - The majority of couple prefer restaurants with an approximate cost of 300 rupees

# Qus 5 - Which mode (online or offline) has received the maximum rating?

```python
dataframe.head()
```

```python
plt.figure(figsize = (6,6))
sns.boxplot(x = 'online_order', y= 'rate', data = dataframe)
```

# conclusion - Offline order received lower ratings in comparision to online order

# Qus 6 - Which type of restaurant received more offline orders, so that Zomato can provide those customers with some good offers?

```python
dataframe.head()
```

```python
pivot_table = dataframe.pivot_table(index='listed_in(type)', columns='online_order', aggfunc='size', fill_value=0)
sns.heatmap(pivot_table, annot=True, cmap="YlGnBu", fmt='d')
plt.title("Heatmap")
plt.xlabel("Online Order")
plt.ylabel("Listed In Type")
plt.show()
```

# Conclusion - Dining restaurants primarily accept offline orders, whereas cafes primarily receive online orders. This suggests that clients prefer orders in person at restaurants, but prefer online orders at cafes.
